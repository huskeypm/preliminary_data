## Constructing Runner Function to Spit out Contour
Following function returns binary image (1 for contour of interest, 0 for background)

```python
def GetContour(image,name,resize=False,saveSegmentedIMG=False):
    if resize:
        height, width = image.shape[:2]
        resizeFactor = 2
        newHeight, newWidth = height * resizeFactor, width * resizeFactor
        image = cv2.resize(image, (newWidth, newHeight), interpolation = cv2.INTER_AREA)
    
    #### Convert to gray scale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    #### Blurring
    kernelsize = 3 # small kernal size means the image becomes sharper
    gray_blur = cv2.GaussianBlur(gray, (kernelsize, kernelsize), 0)
    
    #### Thresholding
    thinningParameter = 3 # larger is more thin
    blockSize = 13 #Size of pixel neighborhood used to calculate threshold value
    retval, thresh = cv2.threshold(gray_blur, 17, 1, cv2.THRESH_BINARY)
    
    #### Resize again
    resize = False
    if resize:
        height, width = thresh.shape[:2]
        resizeFactor = 2
        newHeight, newWidth = height * resizeFactor, width * resizeFactor
        thresh = cv2.resize(thresh, (newWidth, newHeight), interpolation = cv2.INTER_AREA)
    
    #### Closing contours
    numIters = 1
    kernel = np.ones((3, 3), np.uint8)
    closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE,
                               kernel, iterations=numIters)
    
    if saveSegmentedIMG:
        cv2.imwrite("./Segmented_Images/"+name+"_segmented"+fileType,closing)
        print "Wrote "+"./Segmented_Images/"+name+"_segmented"+fileType
    
    #### Gray scale image with background set to pix intensity = 0
    segmented = closing * gray
    
    #### Finding contour
    contours, hierarchy = cv2.findContours(closing,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    perim = []
    for i,cnt in enumerate(contours):
        perim_i = cv2.arcLength(cnt,True)
        perim.append(perim_i)
    maxPerimIdx = np.argmax(perim)
    contours = contours[maxPerimIdx]
    
    return thresh, gray, segmented, contours
    
    
```

## Construction of Radial Histogram
Need to mark centroid for this. Using image moment of segmented (binary) image

```python
def radial_profile(data, center):
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    r = r.astype(np.int)
    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile 
```

```python
def radial_histogram(segmentedImage, contour, name, saveData=False):
    # Finding moments/centroids
    M = cv2.moments(contour)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    center = cx,cy
    
    # Finding radial profile
    radialProfile = radial_profile(segmentedImage,center)
    
    # Plotting/saving of data
    plt.plot(radialProfile)
    plt.title("Radial Histogram of Pixel Intensity for "+name)
    plt.xlabel("Distance from Centroid")
    plt.ylabel("Pixel Intensity")
    if saveData:
        savefig("./histogram_"+name+fileType)
        print "Wrote " + "./histogram_"+name+fileType
    return radialProfile

    
```

```python
radialProfile1 = radial_histogram(segmentedIMG, contour1, fileName, saveData=True)
```
