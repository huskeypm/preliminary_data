```python
import numpy
from numpy import *
import numpy as N
import pylab as P
import numpy as np
diffusion = np.array([[1,2,3,4,6,7],[5.92269E-08, 5.77815E-08, 3.70681E-08,4.28487E-08, 5.41665E-08, 5.02481E-08 ]])
friction = np.array([[1,2,3,4,6,7],[6.99335E-14, 7.1683E-14, 1.11739E-13, 9.66644E-14,7.64669E-14 ,8.24298E-14 ]])
Dbulk = 3.83145E-08
fbulk = 1.08104E-13
x = diffusion[0,:]
x_prime = friction[0,:]
D = diffusion[1,:]
f = friction[1,:]
P.plot(x,D)
P.xlabel('distance to surface[A]')
P.ylabel('diffusion[m^2/sec]')
P.show()
P.plot(x_prime,f)
P.xlabel('distance to surface[A]')
P.ylabel('friction profile')
P.show()
```

```python
Dn = D/Dbulk
P.plot(x,Dn)
P.xlabel('distance to surface[A]')
P.ylabel('D / Dbulk')
P.show()
```

```python
fn = f/fbulk
P.plot(x,fn)
P.xlabel('distance to surface[A]')
P.ylabel('friction / fbulk')
P.show()
```
