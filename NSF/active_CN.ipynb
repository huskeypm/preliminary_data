Notebook that utilizes 3 compartment ODE to show the dephosphorylation of a substrate via activated CN 

### From the mouth of "the man" himself:

- three compartments, Omega1,2 and 3, aligned in sequential order (e.g. Omega1 and 3 are not connected, 1-2, 2-3 are)
- four mobile species, pNPP, NP, PI and Ca, all of which can diffuse between the Omegas (can assume the same diff. coefficient for all).
- four immobile species restricted to Omega 2, CaM, CaCaM, CaCN, CaCaMCN
- Fluxes of the mobile species, Mi, are governed by
  - dMi_dt = -D[ Mi(Omega_i) - Mi(Omega_j) ]    # update based on connectivity of the compartments
  - dCa_dt(Omega1) = f(t)   # TBD time-dependent function
- The reactions occurring in Omega 2 are the following:
  - Ca + CaM <--> CaCaM
  - CaCaM + CaCN <--> CaCaMCN
  - pNPP + CaCaMCN --> NP + PI   + CaCaMCN  
- You can reduce the number of independent variables by replacing CaM and CaCN based on these constraints:
  - CaM_tot = CaM + CaCaM + CaCaMCN
  - CN_tot   = CaCN + CaCaMCN

### Make a Parameter Storage

```python
class empty:
    pass
params = empty()
```

### Define Compartments

```python
dim1 = [50, 50, 50] # [cm]
Omega1 = np.product(dim1) # [cm^3]
params.volume_Omega1 = Omega1

dim2 = [50, 50, 50] # [cm]
Omega2 = np.product(dim2) # [cm^3]
params.volume_Omega2 = Omega2

dim3 = [50, 50, 50] # [cm]
Omega3 = np.product(dim3) # [cm^3]
params.volume_Omega3 = Omega3
```

### Define Diffusion Coefficients for Mobile Species

Based off of Ca diffusion coefficient in cytoplasm paper: https://www.ncbi.nlm.nih.gov/pubmed/3435913

```python
params.D_pNPP = 5.3 * 10**-6 # [cm^2/s]
params.D_NP = 5.3 * 10**-6 # [cm^2/s]
params.D_PI = 5.3 * 10**-6 # [cm^2/s]
params.D_Ca = 5.3 * 10**-6 # [cm^2/s]
```

### Define Reactions

```python
def Rx1(conc_Ca, 
        conc_CaM, 
        conc_CaCaM, 
        kOn,
        kOff):
    delta_Ca = -kOn * conc_Ca * conc_CaM + kOff * conc_CaCaM
    delta_CaM = delta_Ca
    delta_CaCaM = -delta_Ca
    changes = [delta_Ca, delta_CaM, delta_CaCaM]
    return changes

def Rx2(conc_CaCaM, 
        conc_CaCN, 
        conc_CaCaMCN,
        kOn,
        kOff):
    delta_CaCaM = -kOn * conc_CaCaM * conc_CaCN + kOff * conc_CaCaMCN
    delta_CaCN = delta_CaCaM
    delta_CaCaMCN = -delta_CaCaM
    changes = [delta_CaCaM, delta_CaCN, delta_CaCaMCN]
    return changes

def Rx3(conc_pNPP,
        conc_CaCaMCN,
        conc_NP,
        conc_PI,
        kOn):
    delta_pNPP = -kOn * conc_pNPP * conc_CaCaMCN
    delta_CaCaMCN = 0.
    delta_NP = -delta_pNPP
    delta_PI = delta_NP
    changes = [delta_pNPP, delta_CaCaMCN, delta_NP, delta_PI]
    return changes
```

### Define Reaction Rates

```python
params.kOn_Rx1 = .01 # [lazy units]
params.kOff_Rx1 = 0.0001 # [lazy units]

params.kOn_Rx2 = .01 # [lazy units]
params.kOff_Rx2 = 0.0001 # [lazy units]

params.kOn_Rx3 = .01 # [lazy units]
```

### Define Function for Diffusion of Mobile Species

```python
def diffusionOfMobileSpecies(speciesConcInSpace1,speciesConcInSpace2,diffusionConstant):
    return -diffusionConstant * (speciesConcInSpace1 - speciesConcInSpace2)
```

### Define Time Scale and Initial Concentrations

```python
t = np.linspace(0.,1000.,200)

CaM_total_2 = 2000. # [uMol] (arbitrary)
CaCN_total_2  = 2000. # [uMol] (arbitrary)

init_pNPP_1 = 2000. # [uMol] (arbitrary)
init_NP_1 = 2000. # [uMol] (arbitrary)
init_PI_1 = 2000. # [uMol] (arbitrary)
init_Ca_1 = 2000. # [uMol] (arbitrary)

init_pNPP_2 = 2000. # [uMol] (arbitrary)
init_NP_2 = 2000. # [uMol] (arbitrary)
init_PI_2 = 2000. # [uMol] (arbitrary)
init_Ca_2 = 2000. # [uMol] (arbitrary)

init_pNPP_3 = 2000. # [uMol] (arbitrary)
init_NP_3 = 2000. # [uMol] (arbitrary)
init_PI_3 = 2000. # [uMol] (arbitrary)
init_Ca_3 = 2000. # [uMol] (arbitrary)

init_CaCaM_2 = 0. # [uMol] (arbitrary)
init_CaCaMCN_2 = 0. # [uMol] (arbitrary)

initial_concentrations = [init_pNPP_1,
                          init_NP_1,
                          init_PI_1,
                          init_Ca_1,
                          init_pNPP_2,
                          init_NP_2,
                          init_PI_2,
                          init_Ca_2,
                          init_pNPP_3,
                          init_NP_3,
                          init_PI_3,
                          init_Ca_3,
                          init_CaCaM_2,
                          init_CaCaMCN_2]
```

### Define Integrator Function

```python
def updater(yi, t, params):
    ### Pull out your concentrations from the list
    init_pNPP_1 = yi[0]
    init_NP_1 = yi[1]
    init_PI_1 = yi[2]
    init_Ca_1 = yi[3]
    init_pNPP_2 = yi[4]
    init_NP_2 = yi[5]
    init_PI_2 = yi[6]
    init_Ca_2 = yi[7]
    init_pNPP_3 = yi[8]
    init_NP_3 = yi[9]
    init_PI_3 = yi[10]
    init_Ca_3 = yi[11]
    init_CaCaM_2 = yi[12]
    init_CaCaMCN_2 = yi[13]
    
    ### calculate free CaM and CaCN
    init_CaM_2 = CaM_total_2 - init_CaCaM_2 - init_CaCaMCN_2
    init_CaCN_2 = CaCN_total_2 - init_CaCaMCN_2
    
    ### find change in terms due to reaction
    change_Ca_Rx1, change_CaM_Rx1, change_CaCaM_Rx1 = Rx1(init_Ca_2, 
                                                          init_CaM_2, 
                                                          init_CaCaM_2, 
                                                          params.kOn_Rx1,
                                                          params.kOff_Rx1)
    change_CaCaM_Rx2, change_CaCN_Rx2, change_CaCaMCN_Rx2 = Rx2(init_CaCaM_2,
                                                                init_CaCN_2,
                                                                init_CaCaMCN_2,
                                                                params.kOn_Rx2,
                                                                params.kOff_Rx2)
    change_pNPP_Rx3, change_CaCaMCN_Rx3, change_NP_Rx3, change_PI_Rx3 = Rx3(init_pNPP_2,
                                                                            init_CaCaMCN_2,
                                                                            init_NP_2,
                                                                            init_PI_2,
                                                                            params.kOn_Rx3)
    
    ### Find change in terms due to diffusion
    change_pNPP_diff_1to2 = diffusionOfMobileSpecies(init_pNPP_1,init_pNPP_2,params.D_pNPP)
    change_pNPP_diff_2to3 = diffusionOfMobileSpecies(init_pNPP_2,init_pNPP_3,params.D_pNPP)
    change_NP_diff_1to2 = diffusionOfMobileSpecies(init_NP_1,init_NP_2,params.D_NP)
    change_NP_diff_2to3 = diffusionOfMobileSpecies(init_NP_2,init_NP_3,params.D_NP)
    change_PI_diff_1to2 = diffusionOfMobileSpecies(init_PI_1,init_PI_2,params.D_PI)
    change_PI_diff_2to3 = diffusionOfMobileSpecies(init_PI_2,init_PI_3,params.D_PI)
    change_Ca_diff_1to2 = diffusionOfMobileSpecies(init_Ca_1,init_Ca_2,params.D_Ca)
    change_Ca_diff_2to3 = diffusionOfMobileSpecies(init_Ca_2,init_Ca_3,params.D_Ca)
    
    ### Update Terms
    new_pNPP_1 = change_pNPP_diff_1to2 
    new_NP_1 = change_NP_diff_1to2
    new_PI_1 = change_PI_diff_1to2
    new_Ca_1 = change_Ca_diff_1to2
    new_pNPP_2 = -change_pNPP_diff_1to2 + change_pNPP_diff_2to3 + change_pNPP_Rx3
    new_NP_2 = -change_NP_diff_1to2 + change_NP_diff_2to3 + change_NP_Rx3
    new_PI_2 = -change_PI_diff_1to2 + change_PI_diff_2to3 + change_PI_Rx3
    new_Ca_2 = -change_Ca_diff_1to2 + change_Ca_diff_2to3 + change_Ca_Rx1
    new_pNPP_3 = -change_pNPP_diff_2to3
    new_NP_3 = -change_NP_diff_2to3
    new_PI_3 = -change_PI_diff_2to3
    new_Ca_3 = -change_Ca_diff_2to3
    new_CaCaM_2 = change_CaCaM_Rx1 + change_CaCaM_Rx2
    new_CaCaMCN_2 = change_CaCaMCN_Rx2 + change_CaCaMCN_Rx3
    
    updatedY = [new_pNPP_1,
                new_NP_1,
                new_PI_1,
                new_Ca_1,
                new_pNPP_2,
                new_NP_2,
                new_PI_2,
                new_Ca_2,
                new_pNPP_3,
                new_NP_3,
                new_PI_3,
                new_Ca_3,
                new_CaCaM_2,
                new_CaCaMCN_2]
    
    return updatedY
```

### Define Conservation Checker Function

```python
def checkConservation(yi):
    pNPP_1 = yi[:,0]
    NP_1 = yi[:,1]
    PI_1 = yi[:,2]
    Ca_1 = yi[:,3]
    pNPP_2 = yi[:,4]
    NP_2 = yi[:,5]
    PI_2 = yi[:,6]
    Ca_2 = yi[:,7]
    pNPP_3 = yi[:,8]
    NP_3 = yi[:,9]
    PI_3 = yi[:,10]
    Ca_3 = yi[:,11]
    CaCaM_2 = yi[:,12]
    CaCaMCN_2 = yi[:,13]
    
    total_Ca = np.sum(np.stack((Ca_1, Ca_2, Ca_3,CaCaM_2,CaCaMCN_2),axis=1),axis=1)
    deviance_Ca = total_Ca - total_Ca[0]
    if np.max(deviance_Ca) > 1.:
        print "WARNING: Calcium NOT conserved"
    
    #total_pNPP = np.sum(np.stack((pNPP_1,pNPP_2,pNPP_3),axis=1),axis=1)
    #deviance_pNPP = total_pNPP - total_pNPP[0]
    #if np.max(deviance_pNPP) > 1.:
    #    print "WARNING: pNPP NOT conserved"
        
    #total_NPP = np.sum(np.stack((pNPP_1,pNPP_2,pNPP_3),axis=1),axis=1)
    #deviance_pNPP = total_pNPP - total_pNPP[0]
    #if np.max(deviance_pNPP) > 1.:
    #    print "WARNING: pNPP NOT conserved"
```

```python
concentrations = odeint(updater,initial_concentrations,t,args=(params,))

### check for conservation of mass
checkConservation(concentrations)
```
